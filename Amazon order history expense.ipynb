```python
# Libraries needed to import
import pandas as pd
import matplotlib as plt
import datetime
```

```python
 df=pd.read_csv('amazon_order_history.csv') #reads csv file
```

```python
df #prints the dataframe
```

```python
df.fillna(0) #remove NAN and fill with 0
```

```python
df["total"].sum() #total amount spent on the amazon
```

```python
df["total"].max() #maximum amount spent on the order
```

```python
df["total"].min() #minimum amount order
```

```python
df.plot.bar(x='date', y='total', rot=90, figsize=(20,10)) #graph tp show spending over the time
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month #extract month from order date
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year #extract month from order date
df.head()
```

```python
yearly_order=df.groupby('year').sum()["total"]#calculates the amount spent per year
```

```python
yearly_order
```

```python
yearly_order_by_month=df.groupby(['year','month']).sum()["total"] #money spent on amazon monthly year wise 
```
